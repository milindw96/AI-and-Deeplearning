```python
import pytesseract
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
import skimage.data as data
from skimage import color
from skimage import filters
```

```python
%matplotlib inline
%config inlineBackend.figure_format = 'retina'
```

```python
image_name='images.jpeg'
def get_image(image_name):
    """
    image_name (str): name of the name to be processed 
    """
    img_array=io.imread(image_name)
    img_array=color.rgb2gray(img_array)
    return img_array
```

```python
def image_show(image, nrows=1, ncols=1, cmap='gray', **kwargs):
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16, 16))
    ax.imshow(image, cmap='gray') 
    ax.axis('off')
    return fig, ax
```

```python
# set the threshold
# img_array=get_image('images.jpeg')9
img_array=data.page()
# image_threshold = filters.threshold_li(img_array)
image_threshold = filters.threshold_local(img_array, block_size=49)
# enter the image name here 
fig , ax =image_show(img_array > image_threshold)
```

```python
# save the processed image and show the result

fig.savefig('processed_image.png',bbox_inches='tight')
proc_img_array= io.imread('processed_image.png')
img_text =  pytesseract.image_to_string(proc_img_array)
print(img_text)
import os 
os.system('rm -rf processed_image.png')
```
